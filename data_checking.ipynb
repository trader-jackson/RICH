```python
import os
import pandas as pd

def concat_all_csvs(directory):
    csv_files = [
        os.path.join(root, file)
        for root, _, files in os.walk(directory)
        for file in files if file.endswith('.csv')
    ]
    df_list = [pd.read_csv(file) for file in csv_files]
    return pd.concat(df_list, ignore_index=True)

# Set your base directory containing the CSV files
base_dir = "data\\us_stocks\dow30\\raw_2007-01-01-2024-12-29"  # adjust this to your folder
all_data = concat_all_csvs(base_dir)

all_data
```

```python
all_data.groupby("ticker")["date"].count().unique()
```
